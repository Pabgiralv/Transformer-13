```python
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
from torch.utils.data import sampler
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import torch.utils.data as data
from torchvision import datasets
import pandas as pd
import glob
from pandas.core.common import flatten
from PIL import Image
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
mean = [0.4704, 0.4669, 0.3898]
std = [0.2035, 0.2001, 0.2047]
number_of_images = 58388

train_dataset_path = '../input/iais22-birds/birds/birds'
train_transforms = transforms.Compose ([
    transforms.Resize((224,224)),
    transforms.RandomHorizontalFlip(), #Para que los datos no estén sesgados, le aplicamos modificaciones
    transforms.RandomRotation(15),
    transforms.ToTensor(),
    transforms.Normalize(torch.Tensor(mean), torch.Tensor(std))
])
```

```python
orig_set = torchvision.datasets.ImageFolder(root = train_dataset_path, transform = train_transforms)
```

```python
n = len(orig_set) 
n_test = int(0.2 * n)  
train_set_size = int(n * 0.8)
valid_set_size = n - train_set_size
train_set, valid_set = data.random_split(orig_set, [train_set_size, valid_set_size]) #Dividimos los Datasets en 2 de forma aleatoria


train_loader = torch.utils.data.DataLoader(train_set, batch_size = 64, shuffle=True, pin_memory=True, drop_last=True)
valid_loader = torch.utils.data.DataLoader(valid_set, batch_size = 64, shuffle=False, pin_memory=True, drop_last=True)
full_loader = torch.utils.data.DataLoader(orig_set, batch_size = 32, shuffle=True, pin_memory=True, drop_last=True)
```

```python
#Ver si se puede usar la GPU
if torch.cuda.is_available():
    device = torch.device('cuda')
else:
    device = torch.device('cpu')
print(device)
```

```python
def accuracy(model, loader):
    num_correct = 0
    num_total = 0
    model.eval()   #Poner al modelo en modo evaluación, desactiva capas de entrenamiento
    model = model.to(device=device)
    with torch.no_grad():
        for xi, yi in loader:
            xi = xi.to(device=device, dtype = torch.float32)   #Las imágenes son cargadas como tensores de tipo float32
            yi = yi.to(device=device, dtype = torch.long)   #Las etiquetas son cargadas como tensores de tipo long
            scores = model(xi) 
            _, pred = scores.max(dim=1)
            num_correct += (pred == yi).sum() 
            num_total += pred.size(0)
        return float(num_correct)/num_total   
```

```python
def train(model, optimiser, epochs):
    model = model.to(device=device)
    for epoch in range(epochs):
        for i, (xi, yi) in enumerate(train_loader):
            model.train()
            xi = xi.to(device=device, dtype=torch.float32)
            yi = yi.to(device=device, dtype=torch.long)
            scores = model(xi)
            cost = F.cross_entropy(input= scores, target=yi)
            optimiser.zero_grad()           
            cost.backward()
            optimiser.step()              
        acc = accuracy(model, train_loader) 
        val_acc = accuracy(model, valid_loader) 
        date = datetime.now()
        print(f'Epoch: {epoch}, costo: {cost.item()}, accuracy: {acc}, val_accuracy: {val_acc}, hora: {date}')
```

```python
class CNN_model(nn.Module):
    def __init__(self, in_channel):
        super().__init__()
        self.conv1 = nn.Conv2d(in_channels = in_channel, out_channels = 32,
                              kernel_size=3, padding = 1)
        self.conv2 = nn.Conv2d(in_channels = 32, out_channels = 64,
                              kernel_size=3, padding=1)
        self.max_pool1 = nn.MaxPool2d(2,2)

        self.conv3 = nn.Conv2d(in_channels = 64, out_channels = 128,
                              kernel_size=3, padding = 1)
        self.max_pool2 = nn.MaxPool2d(2,2)



        self.fc = nn.Linear(in_features=401408, out_features=400) #El 32 es dependiendo del tamaño de la image
        self.flatten = nn.Flatten()
    def forward(self, x):
        x = F.relu(self.conv2(F.relu(self.conv1(x))))
        x = self.max_pool1(x)
        x = F.relu(self.conv3(x))
        x = self.max_pool2(x)
        x = self.flatten(x)
        return self.fc(x)
```

```python
epochs = 6
lr = 0.00005
modelCNN2 = CNN_model(3)
optimiser = torch.optim.Adam(modelCNN2.parameters(), lr)
```

```python
train(modelCNN2, optimiser, epochs)
```

```python
train(modelCNN2, optimiser, 3)
```

```python
train(modelCNN2, optimiser, 1)
```

```python
train(modelCNN2, optimiser, 2)
```

```python
#Lectura de Submission Test
test_path = '../input/iais22-birds/submission_test/submission_test'
test_images_paths = []
for data_path in glob.glob(test_path + '/*'):
    test_images_paths.append(glob.glob(data_path))

test_images_paths = list(flatten(test_images_paths))
test_images_paths[1]
```

```python
#Función para mover a GPU si es posible
def to_device(data, device):
    "Move data to the device"
    if isinstance(data,(list,tuple)):
        return [to_device(x,device) for x in data]
    return data.to(device,non_blocking = True)


train_transforms = transforms.Compose ([
    transforms.Resize((224,224)),
    transforms.RandomHorizontalFlip(), #Para que no los datos no estén sesgados, le aplicamos modificaciones
    transforms.RandomRotation(15),
    transforms.ToTensor(),
    transforms.Normalize(torch.Tensor(mean), torch.Tensor(std))
])

#Función para predecir imágenes
def predict_img_class(img,model,train_transforms):
    """ Predict the class of image and Return Predicted Class"""
    img = train_transforms(img)
    img = to_device(img.unsqueeze(0), device)
    model.eval()
    prediction =  model(img)
    _, preds = torch.max(prediction, dim = 1)
    return orig_set.classes[preds[0].item()]
```

```python
#Bucle para predecir todas las imágenes y guardarlas en dos listas

Categories = []
Id = []

for i in range (0, len(test_images_paths)):
    img_name = test_images_paths[i]
    x = img_name.replace('../input/iais22-birds/submission_test/submission_test/', '')
    x = x.split('.')
    Id.append(x[0])
    class_predict = predict_img_class(Image.open(img_name),modelCNN2,train_transforms)
    Categories.append(class_predict)
```

```python
#Creación del csv para el submission test

dict = {"Id": Id, "Category": Categories}

df = pd.DataFrame(dict)

df.to_csv('shows-bueno.csv',index=False)
```
